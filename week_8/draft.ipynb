```python
import tt
from tt.amen import amen_solve
```

```python
d = 5
n = 2**d
delta_x = 1 / n
print(delta_x)
```

```python
# second derivative

delta_2 = tt.qlaplace_dd([d])

print(delta_2.full())
delta_2 = delta_2 * (1/(2*delta_x))
print(type(delta_2))
print(delta_2.full())
```

```python
# first derivative 

t = tt.delta(2, d, center = 1)
delta_1 = tt.Toeplitz(t, kind = 'L') - tt.Toeplitz(t, kind = 'L').T
# delta_1 = delta_1.round(1e-14)

print(delta_1.full())
delta_1 = delta_1 * (1/delta_x**2)
print(type(delta_1))
print(delta_1.full())
```

```python
Id = tt.eye(2, d)
Id.full()
```

```python
A = tt.kron(delta_1, Id) + tt.kron(delta_1, delta_2)
# A = A.round(1e-14)
x = tt.ones(2,2*d)
f_0 = tt.rand(n,2*d,r=2)
y = amen_solve(A,f_0,x,1e-6)
```
